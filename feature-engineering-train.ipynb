## Missing Values

### Categorical features

```python
# Missing values in categorical faetures
categorical_nan = [cfeature for cfeature in data.columns if data[cfeature].isnull().any() and data[cfeature].dtypes=='O']
print('Categorical Features (',len(categorical_nan),') having NaN values-')
#percentage of missing values in every column with missing values
for cfeature in categorical_nan:
    print(cfeature, np.round((data[cfeature].isnull().sum() / data.shape[0])*100, 2), '% missing values')
```

```python
# Replace missing values with a new label
def replace_cat_features(data, categorical_nan):
    df=data.copy()
    df[categorical_nan]=df[categorical_nan].fillna('Missing')
    return df
data = replace_cat_features(data, categorical_nan)
data[categorical_nan].isnull().sum()
```

```python
data.head()
```

### Numerical features

```python
# Numerical features which contain missing values
numerical_nan = [feature for feature in data.columns if data[feature].isnull().sum()>0 and data[feature].dtypes!='O']
```

```python
numerical_nan
```

```python
#percentage of missing values in every column with missing values
for feature in numerical_nan:
    print(feature, np.round((data[feature].isnull().sum() / data.shape[0])*100, 2), '% missing values')
```

```python
# Replacing the numerical missing values
for feature in numerical_nan:
    median_value = data[feature].median()
    
    #create a new feature = feature_nan for every feature to capture NaN values
    data[feature+'nan']=np.where(data[feature].isnull(),1,0) #1 for missing value, 0 for non-missing values
    data[feature].fillna(median_value, inplace=True)
```

### Date-Time Variables

```python
year_feature=['YearBuilt', 'YearRemodAdd', 'GarageYrBlt', 'YrSold']
```

```python
data[year_feature].isnull().sum()
```

```python
data[year_feature].head(10)
```

```python
for feature in ['YearBuilt', 'YearRemodAdd', 'GarageYrBlt']:
    data[feature]=data['YrSold'] - data[feature]
```

```python
num_features = ['LotFrontage', 'LotArea', '1stFlrSF', 'GrLivArea', 'SalePrice']

for feature in num_features:
    data[feature] = np.log(data[feature])
```

## Handling Rare Categorical Features
Remove categorical variables that are present less than 1% of the observations

```python
categorical_features = [feature for feature in data.columns if data[feature].dtypes=='O']
```

```python
categorical_features
```

```python
for feature in categorical_features:
    temp = data.groupby(feature)['SalePrice'].count()/len(data)
    temp_df = temp[temp>0.01].index
    data[feature] = np.where(data[feature].isin(temp_df), data[feature], 'Rare_var')
```

```python
data.head(50)
```

```python
len(categorical_features)
```

```python
data[categorical_features].head(10)
```

```python
for feature in categorical_features:
    data = pd.concat([data, pd.get_dummies(data[feature], prefix=feature, drop_first=True)], axis=1)
```

```python
data.shape
```

```python
data.drop(categorical_features, axis=1, inplace=True)
```

## Feature Scaling

```python
feature_scale = [feature for feature in data.columns if feature not in ['Id', 'SalePrice']]
len(feature_scale)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(data[feature_scale])
```

```python
scaler.transform(data[feature_scale])
```

```python
df = pd.concat([data[['Id', 'SalePrice']].reset_index(drop=True), 
              pd.DataFrame(scaler.transform(data[feature_scale]), columns = feature_scale)], axis=1)
```
