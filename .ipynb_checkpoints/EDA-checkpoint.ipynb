```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
#Display all the columns of the dataset
pd.set_option('display.max_columns', 500)
```

```python
data = pd.read_csv("train.csv")
```

## Missing Values

```python
nan_cols = [features for features in data.columns if data[features].isnull().any()]

#percentage of missing values in every column with missing values
for feature in nan_cols:
    print(feature, np.round((data[feature].isnull().sum() / data.shape[0])*100, 2), '% missing values')
```

```python
# Relationship between missing values and SalePrices
for feature in nan_cols:
    df = data.copy()
    df[feature] = np.where(df[feature].isnull(),1,0) # 1 indicates observation was missing and 0 indicates it was not missing
    df.groupby(feature)['SalePrice'].median().plot.bar(color=['orange','blue'])
    plt.title(feature)
    plt.show()
```

## Numerical columns

```python
# List of numerical columns 
numerical_features = [feature for feature in data.columns if data[feature].dtypes !='O'] # 'O' means object here
print('No. of numerical features: ', len(numerical_features))
```

```python
data[numerical_features].head()
```

## Date-Time columns

```python
# List of columns that contain year info
year_features = [feature for feature in numerical_features if 'Yr' in feature or 'Year' in feature]

year_features
```

```python
# Relation between year the house is sold and SalePrice
plt.figure(figsize=(7,4))
data.groupby('YrSold')['SalePrice'].median().plot()
plt.xlabel('Year Sold')
plt.ylabel('Median House Price')
```

```python
# Relation between all year feature columns with SalePrice
for feature in year_features:
    if feature!='YrSold':
        df=data.copy()
        df[feature] = df['YrSold'] - df[feature]
        plt.scatter(df[feature], df['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('SalePrice')
        plt.show()
```

#### Numercial features are of 2 types:
1. Continuous variable feature
2. Discrete variable feature

### Discrete Features

```python
# List of discrete variable features
discrete_features = [feature for feature in numerical_features if data[feature].nunique() < 25 and feature not in year_features+['Id']]

print('No. of dicrete features: ', len(discrete_features))
discrete_features
```

```python
data[discrete_features].head()
```

```python
# Relation between discrete featues and SalePrice

for feature in discrete_features:
    df = data.copy()
    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.xlabel(feature)
    plt.ylabel('SalePrice')
    plt.show()
```

### Continuous Features

```python
# List of continuous variable features
continuous_features = [feature for feature in numerical_features if feature not in discrete_features+year_features+['Id']]

print('No. of dicrete features: ', len(continuous_features))
continuous_features
```

```python
# relation between continuous features and SalePrice
# Use histograms for continuous features

for feature in continuous_features:
    df = data.copy()
    df[feature].hist(bins=25)
    plt.xlabel(feature)
    plt.ylabel("Count")
    plt.show()
```

In Regression, whenever our data is not normally distributed(visible in above histgrams), or skewed, we transform them into log normal distributions.

```python
for feature in continuous_features:
    df = data.copy()
    if 0 in df[feature].unique():
        pass
    else:
        df[feature] = np.log(df[feature])
        df['SalePrice'] = np.log(df['SalePrice'])
        plt.scatter(df[feature], df['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('SalePrice')
        plt.show()
```

## Outliers

```python
for feature in continuous_features:
    df = data.copy()
    if 0 in df[feature].unique():
        pass
    else:
        df[feature] = np.log(df[feature])
        df.boxplot(column=feature)
        plt.ylabel(feature)
        plt.show()
```

## Categorical features

```python
# List of categorical features
categorical_features = [feature for feature in data.columns if data[feature].dtype=='O']

print('No. of Categorical features: ', len(categorical_features))
categorical_features
```

```python
data[categorical_features].head()
```

```python
for feature in categorical_features:
    print(feature, "has", data[feature].nunique(),"categories")
```

```python
# Relation between Categorical_features and SalePrice

for feature in categorical_features:
    df = data.copy()
    df.groupby(feature)['SalePrice'].median().plot.bar(color=['red','blue','green','orange','magenta','black','brown'])
    plt.xlabel(feature)
    plt.ylabel('SalePrice')
    plt.show()
```
